# Distribution des Types (TV shows and Movies)

```python

import matplotlib.pyplot as plt

# Count the number of occurrences of each unique value in the 'type' column of the DataFrame
counts = data['type'].value_counts()

# Get an array of unique values in the 'type' column of the DataFrame
labels = data['type'].unique()

# Create a pie chart using Seaborn
plt.figure(figsize=(8, 6))
plt.title('Distribution des Types')
sns.set_palette('bright')
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops={'linewidth': 2, 'edgecolor': 'white'})
plt.show()
```

# Top 5 catégories

```python
# Count the number of occurrences of each unique value in the 'listed_in' column of the DataFrame
counts = data['listed_in'].value_counts()

# Get the top 5 categories
top_categories = counts[:5]

# Create a pie chart using Seaborn
plt.figure(figsize=(8, 6))
plt.title('Top 5 Catégories')
sns.set_palette('bright')
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=90, wedgeprops={'linewidth': 2, 'edgecolor': 'white'})
plt.show()
```

# Top 5 réalisateur

# Top 10  acteur

```python

# Split the 'cast' column by comma, explode the resulting list, and strip spaces from the elements
actors = data['cast'].str.split(',', expand=True).stack().str.strip()

# Count the number of occurrences of each unique value in the 'actors' column
counts = actors.value_counts()

# Get the top 10 actors
top_actors = counts[:10]

# Create a circular chart using Seaborn
plt.figure(figsize=(8, 6))
plt.title('Top 10 Acteurs')
sns.set_palette('bright')
plt.pie(top_actors, labels=top_actors.index, autopct='%1.1f%%', startangle=90, counterclock=False, wedgeprops={'linewidth': 2, 'edgecolor': 'white'})
plt.show()
```

# Top 5 Séries avec le plus grand nombre de saisons

```python
# Sélectionner uniquement les séries télévisées
tv_shows = data[data["type"] == "TV Show"]

# Grouper par titre et compter le nombre de saisons pour chaque titre
grouped = tv_shows.groupby("title")["duration"].max().reset_index()

# Supprimer les non-numeric characters de la colonne "duration"
grouped["duration"] = grouped["duration"].str.replace(" Season", "")
grouped["duration"] = grouped["duration"].str.replace("s", "")

# Convertir la colonne "duration" en entiers
grouped["duration"] = grouped["duration"].astype(int)

# Trier par ordre décroissant de nombre de saisons et sélectionner les cinq premiers
top_5 = grouped.sort_values("duration", ascending=False).head(5)

# Créer un circular chart avec Seaborn
plt.figure(figsize=(8, 6))
plt.title('Top 5 Séries avec le plus grand nombre de saisons')
sns.set_palette('bright')
plt.pie(top_5["duration"], labels=top_5["title"], autopct='%1.1f%%', startangle=90, counterclock=False, wedgeprops={'linewidth': 2, 'edgecolor': 'white'})
plt.show()
```

# Contenus similaires

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sélectionner la colonne de texte à utiliser pour la vectorisation
text_column = "description"

# Vectoriser le texte des contenus à l'aide de TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words="english")
vectorized_text = vectorizer.fit_transform(data[text_column])

# Calculer la similarité cosinus entre chaque paire de contenus
similarity_matrix = cosine_similarity(vectorized_text)

# Parcourir la matrice de similarité pour identifier les paires de contenus similaires
threshold = 0.8 # Seuil de similarité à partir duquel deux contenus sont considérés comme similaires
similar_contents = []
for i in range(similarity_matrix.shape[0]):
    for j in range(i+1, similarity_matrix.shape[1]):
        if similarity_matrix[i,j] >= threshold:
            similar_contents.append((i,j))
            print(f"Contenu {i+1} et contenu {j+1} sont similaires.")
            print(f"Ligne {i+1}: {data.loc[i][text_column]}")
            print(f"Ligne {j+1}: {data.loc[j][text_column]}\n")
```

# Numbers of Movies and TV Shows Added by Disney+ in the Last 3 Years

```python
import pandas as pd
import matplotlib.pyplot as plt


# Convert the "date_added" column to a datetime object
data["date_added"] = pd.to_datetime(data["date_added"])

# Select only TV shows and movies
tv_shows = data[data["type"] == "TV Show"]
movies = data[data["type"] == "Movie"]

# Calculate the number of TV shows and movies added each year
tv_shows_count = tv_shows.groupby(tv_shows["date_added"].dt.year)["title"].count()
movies_count = movies.groupby(movies["date_added"].dt.year)["title"].count()

# Plot the results as a graph
plt.figure(figsize=(10,6))
plt.plot(tv_shows_count.index, tv_shows_count.values, label="TV Shows", linewidth=2)
plt.plot(movies_count.index, movies_count.values, label="Movies", linewidth=2)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Number of Productions", fontsize=14)
plt.title("Numbers of Movies and TV Shows Added by Disney+ in the Last 3 Years", fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(alpha=0.5)
plt.legend(fontsize=12)
plt.show()
```
